Creates plots to visualise time "resets" and how this was fixed.

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Loads Excel file
file_path = "/Users/maks/Documents/MSc_project/data/coords_extracted/yellow8.xlsx"
df = pd.read_excel(file_path)

#Plots the "time" column
plt.plot(df["time"])
plt.xlabel("Time-step")
plt.ylabel("Time [seconds]")
plt.title("Example of restes in time data")
plt.grid(True)
plt.show()
```

```python
#Loads Excel file
file_path = "/Users/maks/Documents/MSc_project/data/variables_data/yellow8.xlsx"
df2 = pd.read_excel(file_path)

#Plots the "time" column
plt.plot(df2["time"])
plt.xlabel("Time-step")
plt.ylabel("Time [seconds]")
plt.title("Example plot of a time data column")
plt.grid(True)
plt.show()
```

```python
#Loads two chosen Excel files
file_path1 = "/Users/maks/Documents/MSc_project/data/coords_extracted/yellow8.xlsx"
df1 = pd.read_excel(file_path1)

file_path2 = "/Users/maks/Documents/MSc_project/data/variables_data/yellow8.xlsx"
df2 = pd.read_excel(file_path2)
```

```python
#Creates a figure with two plots, one above the other
plt.figure(figsize=(7, 7))

#First file
plt.subplot(2, 1, 1)
plt.plot(df1["time"])
plt.xlabel("Time-step")
plt.ylabel("Time [seconds]")
plt.title("Example of resets in time data")
plt.grid(True)

#Second file
plt.subplot(2, 1, 2)
plt.plot(df2["time"])
plt.xlabel("Time-step")
plt.ylabel("Time [seconds]")
plt.title("Adjusted time data")
plt.grid(True)


plt.tight_layout()
plt.show()
```
