Creates autocorrelograms for chosen files to see how much past data influences a current time-step. Done on velocity.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = "/Users/maks/Documents/MSc_project/data/variables_data/blue41.xlsx"

df = pd.read_excel(path, usecols = "E:G")

velocity_x = np.array(df["velocity_x"].values)
velocity_y = np.array(df["velocity_y"].values)
velocity_z = np.array(df["velocity_z"].values)

v_x = velocity_x[: 300]
v_y = velocity_y[: 300]
v_z = velocity_z[: 300]

x = pd.plotting.autocorrelation_plot(v_x)

#Limits the range of displayed data
plt.xlim([0, 45])

#Chnages the ticks on the horizontal axis to effectively show lag in seconds
current_ticks = plt.xticks()[0]
new_ticks = current_ticks / 30
plt.xticks(current_ticks, labels=[f'{tick:.2f}' for tick in new_ticks])
plt.xlabel("Lag [seconds]")

# Show the plot
plt.show()
```
